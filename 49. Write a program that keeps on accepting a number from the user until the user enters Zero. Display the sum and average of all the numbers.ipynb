# 49. Write a program that keeps on accepting a number from the user until the user enters Zero. Display the sum and average of all the numbers.

```python
num = int(input("Enter your number: "))
add = 0
avg = 0
count = 0

while True:
    if num != 0:
        add = add + num
        count = count + 1
        avg = add/count
        num = int(input("Another number please: "))
        
    else:
        print("Thank you")
        break
print('Your sum is:', add)
print('Your average is:', avg)
```

```python
sum_nums = 0       # to store sum
count = 0          # to count numbers

while True:
    num = int(input("Enter number (0 to stop): "))
    
    if num == 0:   # stopping condition
        break
    
    sum_nums += num   # add number to sum
    count += 1        # increase count

# Avoid division by zero
if count == 0:
    print("No numbers entered")
else:
    average = sum_nums / count
    print("Sum:", sum_nums)
    print("Average:", average)
```

```python
# Method 2: Using list (alternative)
nums = []

while True:
    num = int(input("Enter number (0 to stop): "))
    if num == 0:
        break
    nums.append(num)

if len(nums) == 0:
    print("No numbers entered")
else:
    print("Sum:", sum(nums))
    print("Average:", sum(nums)/len(nums))
```
