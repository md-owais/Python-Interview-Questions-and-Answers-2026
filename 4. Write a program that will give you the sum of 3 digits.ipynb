4.	Write a program that will give you the sum of 3 digits

```python
num = int(input("Enter three digit number: "))
a = num % 10 #(345 %10 = 5)
num = num // 10 #(345 // 10 = 34)
b = num % 10 #(34 % 10 = 4)
c = num // 10 #(34 // 10 = 3)
result = (a + b + c)
print("The sum of three digit is:", result)
```
