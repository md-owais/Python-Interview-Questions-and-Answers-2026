# 60. Write a python program to find the max item from a list without using the max function

```python
L = [1,2,3,5,6,7]
max = L[0]

for i in L:
    if i > max:
        max = i
print(max)
```

```python
# 1. Using Loop (Most Important – Interview ⭐)
def find_max(lst):
    maximum = lst[0] # assume first element is max
    
    for num in lst:
        if num > maximum:
            maximum = num
            
    return maximum

data = [3, 7, 2, 9, 5]
print(find_max(data))
```

```python
# 2. Using While Loop
def find_max(lst):
    i = 0
    maximum = lst[0]
    
    while i < len(lst):
        if lst[i] > maximum:
            maximum = lst[i]
        i += 1
    
    return maximum

data = [3, 7, 2, 9, 5]
print(find_max(data))
```

```python
# 3. Using Sorting (Not Optimal but Valid)
def find_max(lst):
    lst.sort()
    return lst[-1]

data = [3, 7, 2, 9, 5]
print(find_max(data))
```

```python
# Edge Case Handling (Important 💡)
def find_max(lst):
    if not lst:
        return "List is empty"
    
    maximum = lst[0]
    
    for num in lst:
        if num > maximum:
            maximum = num
    
    return maximum

data = [3, 7, 2, 9, 5]
print(find_max(data))
```
