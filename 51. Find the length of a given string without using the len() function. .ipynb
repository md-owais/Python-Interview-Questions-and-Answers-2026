# 51.	Find the length of a given string without using the len() function. 

```python
a = input("Enter your string: ")
count = 0

for i in a:   # iterate through each character
    count = count + 1   # increase count for each character
print("Length of the string is:", count)
```

```python
# Using while loop
s = input("Enter string: ")

count = 0

while True:
    try:
        s[count] # try accessing character
        count += 1
    except IndexError:
        break   # stops when index out of range
print("Length:", count)
```

```python
# Using recursion (Advanced)
def length(s):
    if s == "":
        return 0
    return 1 + length(s[1:]) # remove first char each time

s = input("Enter string: ")
print("Length:", length(s))
```
