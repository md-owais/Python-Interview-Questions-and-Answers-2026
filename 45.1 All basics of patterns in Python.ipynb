# STEP 2: Basic Star Patterns

```python
n = 4

for i in range(n): # rows
    for j in range(n): # columns
        print("*", end= " ")
    print()
    
# Same number of rows & columns
```

```python
for i in range(1, 5):
    print("*" * i) # i stars
# Stars = row number
```

```python
for i in range(4, 0, -1):
    print("*" * i)
# Reverse loop
```

```python
n = 3

for i in range(1, n + 1):
    print("*" * i)
    
for i in range(n-1, 0, -1):
    print("*" * i)
```
