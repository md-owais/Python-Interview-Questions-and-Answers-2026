# 36. Write a program to find the compound interest 

```python
p = int(input("Enter your principal: "))
r = int(input("Enter rate of interest: "))
t = int(input("Enter the time period elapsed: "))

amount = p*(1+r/100)**t
print("Amount is:", amount)
ci = amount - p
print("Compound Interest is:",ci)
```

```python
# Advanced Version (Compounded multiple times per year)
# Formula: A P(1+R/100*n)^n*t

p = float(input("Principal: "))
r = float(input("Rate: "))
t = float(input("Time: "))
n = int(input("Times compounded per year: "))

amount = p * (1 + r/(100*n)) ** (n*t)
ci = amount - p

print("Compound Interest:", ci)
```
