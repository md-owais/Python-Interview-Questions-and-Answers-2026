# 41.	Write a program to print the following pattern

```python
n = int(input("Enter number of rows: "))

for i in range(1, n + 1): # outer loop controls rows
    for j in range(i): # inner loop prints stars
        print("*", end= " ") # print star with new line
    print()      # move to next line after each row
```

```python
# Method 2: Using string multiplication (Pythonic way)
n = 5

for i in range(1, n + 1):
    print("*" * i)   # prints i number of stars
```
