# 34. Print first 25 prime numbers

```python
counter = 0
num = 2
while (counter < 25):
    for i in range(2, num):
        if num % i == 0:
            break
    else:
        print(num)
        counter += 1
    num = num + 1
```

```python
count = 0
num = 2

while count < 25:
    is_prime = True

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num)
        count += 1

    num += 1
```
