```python
age1 = int(input('Enter the first age:'))
age2 = int(input('Enter the second age:'))
age3 = int(input('Enter the third age:'))
# print(max(age1,age2,age3))
oldest = max(age1,age2,age3) # using max() - Built in Function
print("The oldest age is:", oldest)
```

```python
age1 = int(input('Enter the first age:'))
age2 = int(input('Enter the second age:'))
age3 = int(input('Enter the third age:'))

if age1 >= age2 and age1 >= age3: # using if-elif-else condition
    print("oldest age is:", age1)
elif age2 >= age1 and age2 >= age3:
    print("oldest age is:", age2)
else:
    print("oldest age is:", age3)
```

```python
# Using List + Sort
ages = []
for i in range(3):
    ages.append(int(input("Enter age: ")))
ages.sort()

print("oldest age is:", ages[-1])
```

```python
# Using List + Loop (Manual Maximum)
ages = [age1, age2, age3]

largest = ages[0]
for age in ages:
    if age > largest:
        largest = age
print("Oldest age is:", largest)
```
