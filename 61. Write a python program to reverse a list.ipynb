# 61. Write a python program to reverse a list

```python
L = [1, 2, 3, 4, 5]
rev = []

for i in range(len(L)-1,-1,-1):
    rev.append(L[i])
print(rev)   
```

```python
# 1. Using Slicing (Simple & Best ⭐)
def reverse_list(lst):
    return lst[::-1]

data = [1, 2, 3, 4, 5]
print(reverse_list(data))
```

```python
# 2. Using reverse() (In-place)
def reverse_list(lst):
    lst.reverse()
    return lst

data = [1, 2, 3, 4, 5]
print(reverse_list(data))

# Modifies the original list
```

```python
# 3. Using Loop (Interview Important ⭐)
def reverse_list(lst):
    reversed_list = []
    
    for i in range(len(lst) - 1, -1, -1):
        reversed_list.append(lst[i])
    
    return reversed_list

# Example
data = [1, 2, 3, 4, 5]
print(reverse_list(data))
```

```python
# 4. Using reversed()
def reverse_list(lst):
    return list(reversed(lst))

data = [1, 2, 3, 4, 5]
print(reverse_list(data))
```
