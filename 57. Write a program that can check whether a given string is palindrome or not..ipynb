# 57. Write a program that can check whether a given string is palindrome or not.

```python
s = input("Enter String: ")
rev = ''

for i in range(len(s)-1,-1,-1):
    rev = rev + s[i]
print(rev)
if rev == s:
    print("Palindrome")
else:
    print("Not a Palindome")
```

```python
# Using String Slicing (Most Recommended)
def is_palindrome(s):
    s = s.lower() # optional: ignore case
    return s == s[::-1]

#example
s = 'Madam'
if is_palindrome(s):
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# Using Loop (Interview Friendly Logic)
def is_palindrome(s):
    s = s.lower()
    n = len(s)
    
    for i in range(n // 2):
        if s[i] != s[n - i - 1]:
            return False
    return True

# Example
print(is_palindrome("racecar"))
```

```python
# Ignore Spaces & Special Characters
import re

def is_palindrome(s):
    s = re.sub(r'[^a-zA-Z0-9]', '', s).lower()
    return s == s[::-1]

# Example
print(is_palindrome("A man, a plan, a canal: Panama"))
```

```python
# Using While Loop
def is_palindrome(s):
    s = s.lower()
    left, right = 0, len(s) - 1
    
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    
    return True
print(is_palindrome("racecar"))
```
