# 66. Write a program to check if a list is in ascending order or not

```python
L = [1,2,3,4,5]
flag = 0

for i in range(0,len(L)-1):
    if L[i] >= L[i+1]:
        flag = 1
        break
if flag == 0:
    print('Sorted')
else:
    print('Not Sorted')
```

```python
# 1. Using Loop (Most Important ⭐)
def is_ascending(lst):
    for i in range(len(lst) - 1):
        if lst[i] > lst[i + 1]:
            return False
    return True

# Example
data = [1, 2, 3, 4, 5]
print(is_ascending(data))
```

```python
# 2. Using all() (Pythonic Way ⭐)
def is_ascending(lst):
    return all(lst[i] <= lst[i+1] for i in range(len(lst)-1))

# Example
print(is_ascending([1, 2, 2, 3]))

# all() returns True only if all comparisons are true
```

```python
# 3. Compare with Sorted List
def is_ascending(lst):
    return lst == sorted(lst)

print(is_ascending([1,2,3,4]))

# Not optimal for interviews (extra space + sorting)
```

```python
# 4. Strictly Increasing (No Duplicates Allowed)
def is_strictly_ascending(lst):
    return all(lst[i] < lst[i+1] for i in range(len(lst)-1))

print(is_strictly_ascending([1,2,3,4]))
```
