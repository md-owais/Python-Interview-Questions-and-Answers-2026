```python
# Method 1: Using nested loops (Best for understanding)
n = 4   # number of rows
num = 1 # starting number

for i in range(1, n + 1):      # controls rows
    
    for j in range(i):         # prints i numbers in each row
        print(num, end=" ")
        num += 1               # increment number after printing
    
    print()   # move to next line
```

```python
# Method 2: Cleaner version
n = 4
num = 1

for i in range(1, n + 1):
    row = []
    
    for j in range(i):
        row.append(num)
        num += 1
    
    print(*row)
```
