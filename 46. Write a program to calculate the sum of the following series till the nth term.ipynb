# 46.	Write a program to calculate the sum of the following series till the nth term
1/1! + 2/2! + 3/3! + 4/4! +…….+ n/n!

n will be provided by the user


```python
n = int(input("Enter value of n: "))

fact = 1      # to store factorial
sum_series = 0

for i in range(1, n + 1):
    fact = fact * i          # calculate i!
    term = i / fact          # calculate i / i!
    sum_series += term       # add to total sum

print("Sum of series:", sum_series)
```

```python
# Method 2: Using math factorial
import math

n = int(input("Enter n: "))
sum_series = 0

for i in range(1, n + 1):
    sum_series += i / math.factorial(i)

print("Sum:", sum_series)
```
