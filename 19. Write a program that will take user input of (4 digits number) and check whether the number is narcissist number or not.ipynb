# 19.	Write a program that will take user input of (4 digits number) and check whether the number is narcissist number or not.

1634

```python
user_input = int(input("Enter 4-digit number: "))
num = user_input

a = num % 10
num = num // 10

b = num % 10
num = num // 10

c = num % 10
d = num // 10

if (a**4) + (b**4) + (c**4) + (d**4) == user_input:
    print("Narcissistic Number")
else:
    print("Not a Narcissistic Number")
```

```python
num = int(input("Enter a 4-digit number: "))

temp = num
n = 4   # since it's a 4-digit number
result = 0

while temp > 0:
    digit = temp % 10
    result += digit ** n
    temp = temp // 10

if result == num:
    print("Narcissistic Number")
else:
    print("Not a Narcissistic Number")
```

```python
num = input("Enter a 4-digit number: ")

result = sum(int(d)**4 for d in num)

if result == int(num):
    print("Narcissistic Number")
else:
    print("Not Narcissistic Number")
```
