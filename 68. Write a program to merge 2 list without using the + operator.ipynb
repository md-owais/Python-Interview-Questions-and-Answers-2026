# 68.	Write a program to merge 2 list without using the + operator

```python
L1 = [1,2,3,4]
L2 = [5,6,7,8]
L = []

for i in L1:
    L.append(i)
    
for j in L2:
    L.append(j)
print(L)
```

```python
# 1. Using extend() (Best & Simple ⭐)
def merge_lists(lst1, lst2):
    result = lst1.copy()  # avoid modifying original list
    result.extend(lst2)
    return result

a = [1,2,3,4]
b = [5,6,7,8]
print(merge_lists(a, b))    
```

```python
# 2. Using Loop (Interview Logic ⭐)
def merge_lists(lst1, lst2):
    result = []
    
    for item in lst1:
        result.append(item)
    
    for item in lst2:
        result.append(item)
    
    return result

a = [1,2,3,4]
b = [5,6,7,8]
print(merge_lists(a, b))
```

```python
# 3. Using List Comprehension
def merge_lists(lst1, lst2):
    return [item for lst in (lst1, lst2) for item in lst]

a = [1,2,3,4]
b = [5,6,7,8]
print(merge_lists(a, b))
```

```python
# 4. Using itertools.chain() (Advanced Pythonic ⭐)
from itertools import chain

def merge_lists(lst1, lst2):
    return list(chain(lst1, lst2))

a = [1,2,3,4]
b = [5,6,7,8]
print(merge_lists(a, b))
```

```python
# 5. Using Unpacking (Modern Python ⭐)
def merge_lists(lst1, lst2):
    return [*lst1, *lst2]
a = [1,2,3,4]
b = [5,6,7,8]
print(merge_lists(a, b))

#No + used, clean and efficient
```
