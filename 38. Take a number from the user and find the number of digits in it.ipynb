# 38.	Take a number from the user and find the number of digits in it. 

```python
a = int(input("Enter a number: "))
count = 0 

while(a > 0):
    a = a // 10
    count += 1
print("Number of digits:", count)
```

```python
num = int(input("Enter a number: "))

count = 0

if num == 0:
    count = 1
else:
    num = abs(num)   # handles negative numbers
    while num > 0:
        count += 1
        num //= 10

print("Number of digits:", count)
```

```python
num = input("Enter a number: ")

print("Number of digits:", len(num.lstrip('-')))
```

```python
import math

num = int(input("Enter a number: "))

if num == 0:
    print("Number of digits: 1")
else:
    digits = int(math.log10(abs(num))) + 1
    print("Number of digits:", digits)
```
