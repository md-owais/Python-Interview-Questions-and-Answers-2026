# 11.	Write a program to find the simple interest when the value of principle, rate of interest and time period is given.
Interest rate(r) = ( Simple Interest(I)*100)/(Principle(P)*Time(T))
Where:

I = Simple Interest earned

P = Principal amount (the initial sum of money)

T = Time period (in years)

```python
P = int(input("Enter your Pricncipal Amount: "))
R = int(input("Enter your Interest Rate: "))
T = int(input("Enter your time period in years: "))

SI = (P * R * T)/ 100

print("Your Simple Interest is:", SI)
```
