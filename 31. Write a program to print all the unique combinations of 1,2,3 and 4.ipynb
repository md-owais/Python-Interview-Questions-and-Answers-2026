# 31.	Write a program to print all the unique combinations of 1,2,3 and 4

```python
for i in range(1,5):
    for j in range(1,5):
        if(i != j):
            print(i, j)
```

```python
# Using itertools.permutations
# 4! = 24
from itertools import permutations

nums = [1, 2, 3, 4]

for p in permutations(nums):
    print(*p)
```

```python
nums = [1, 2, 3, 4]

for i in nums:
    for j in nums:
        for k in nums:
            for l in nums:
                if i != j and i != k and i != l and j != k and j != l and k != l:
                    print(i, j, k, l)
```

```python
from itertools import permutations

nums = [1, 2, 3, 4]

for p in permutations(nums):
    print("".join(map(str, p)))
```

```python
from itertools import combinations

nums = [1, 2, 3, 4]

for r in range(1, len(nums) + 1):
    for c in combinations(nums, r):
        print(*c)
```

```python
#What you should write in exam/interview
#If the question literally says combination, then the safer correct answer is:
from itertools import combinations

nums = [1, 2, 3, 4]

for r in range(1, len(nums) + 1):
    for c in combinations(nums, r):
        print(c)
```
