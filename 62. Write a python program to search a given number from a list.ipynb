# 62. Write a python program to search a given number from a list

```python
L = [1, 2, 3, 4, 5, 6, 7, 8, 9]
num = int(input('Enter your number: '))

for i in L:
    if i == num:
        print('True')
        break
else:
    print(False)
```

```python
# 1. Linear Search (Most Important – Interview ⭐)
def linear_search(lst, target):
    for i in range(len(lst)):
        if lst[i] == target:
            return i  # return index
    return -1
data = [10, 20, 30, 40, 50]
print(linear_search(data, 30))
```

```python
# 2. Using in Operator (Simple)
def search(lst, target):
    return target in lst

print(search([10, 20, 30], 20))

# Returns True or False
```

```python
# 3. Using index()
def search(lst, target):
    if target in lst:
        return lst.index(target)
    return -1
print(search([10, 20, 30], 20))
# Returns index (but internally does linear search)
```

```python
# 4. Binary Search (Efficient – Sorted List Required ⭐)
def binary_search(lst, target):
    left, right = 0, len(lst) - 1
    
    while left <= right:
        mid = (left + right) // 2
        
        if lst[mid] == target:
            return mid
        elif lst[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return -1

# Example (sorted list required)
data = [10, 20, 30, 40, 50]
print(binary_search(data, 40))
```
