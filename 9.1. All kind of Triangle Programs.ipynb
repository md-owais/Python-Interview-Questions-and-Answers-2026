# 1. Check if Angles Form a Triangle
Sum of all three angles must be 180°

Each angle must be greater than 0

```python
a = int(input("Enter first angle: "))
b = int(input("Enter second angle: "))
c = int(input("Enter third angle: "))

if a + b + c == 180 and a > 0 and b > 0 and c > 0:
    print("Triangle can be formed")
else:
    print("Triangle cannot be formed")
```

# 2. Check if Sides Form a Triangle

Rule: Sum of any two sides must be greater than the third side

```python
a = int(input("Enter side 1: "))
b = int(input("Enter side 2: "))
c = int(input("Enter side 3: "))

if (a + b > c) and (b + c > a) and (a + c > b):
    print("Triangle can be formed")
else:
    print("Triangle cannot be formed")
```

# 3. Find Type of Triangle (Using Sides)
Types

Equilateral → All sides equal

Isosceles → Two sides equal

Scalene → All sides different

```python
a = int(input("Enter side 1: "))
b = int(input("Enter side 2: "))
c = int(input("Enter side 3: "))

if a == b == c:
    print("Equilateral Triangle")
elif a == b or b == c or a == c:
    print("Isosceles Triangle")
else:
    print("Scalene Triangle")
```

# 4. Check Right-Angled Triangle
Rule: Pythagoras theorem

a^2 + b^2 = c^2

# Method 1 – Direct Pythagoras Formula

```python
a = int(input("Enter side a: "))
b = int(input("Enter side b: "))
c = int(input("Enter side c: "))

if a*a + b*b == c*c:
    print("Right angled triangle")
else:
    print("Not a right angled triangle")
```

# Method 2 – Checking All Combinations

(Works even if hypotenuse is not the last value)

```python
a = int(input("Enter side a: "))
b = int(input("Enter side b: "))
c = int(input("Enter side c: "))

if a*a + b*b == c*c or a*a + c*c == b*b or b*b + c*c == a*a:
    print("Right angled triangle")
else:
    print("Not a right angled triangle")
```

# Method 3 – Using Sorting (Best approach)

```python
a = int(input())
b = int(input())
c = int(input())

sides = sorted([a, b, c])

if sides[2]**2 == sides[0]**2 + sides[1]**2:
    print("Right angled triangle")
else:
    print("Not a right angled triangle")
```

# Method 4 – Using Function 

```python
def check_triangle(a,b,c):
    if a*a + b*b == c*c or a*a + c*c == b*b or b*b + c*c == a*a:
        return "Right angled triangle"
    return "Not a right angled triangle"

print(check_triangle(5,12,13))
```

# 5. Triangle Type Using Angles
Types

Acute → All angles < 90

Right → One angle = 90

Obtuse → One angle > 90

```python
a = int(input("Enter angle 1: "))
b = int(input("Enter angle 2: "))
c = int(input("Enter angle 3: "))

if a + b + c == 180:
    if a < 90 and b < 90 and c < 90:
        print("Acute Triangle")
    elif a == 90 or b == 90 or c == 90:
        print("Right Triangle")
    else:
        print("Obtuse Triangle")
else:
    print("Not a valid Triangle")
```
