# 63. Write a program that can create a new list from a given list where each item in the new list is square of the item of the old list

```python
L1 = [2, 3, 4, 5, 6]
L2 = []

for i in L1:
    L2.append(i**2)
print(L2)   
```

```python
# 1. Using Loop (Interview Standard ⭐)
def square_list(lst):
    result = []
    
    for num in lst:
        result.append(num**2)
    return result

data = [1, 2, 3, 4, 5]
print(square_list(data))
```

```python
# 2. Using List Comprehension (Best Pythonic Way ⭐)
def square_list(lst):
    return [num **2 for num in lst]
print(square_list([1, 2, 3, 4]))
```

```python
# 3. Using map() Function
def square_list(lst):
    return list(map(lambda x: x**2, lst))

# Example
print(square_list([1, 2, 3, 4]))
```

```python
# Edge Case Handling
def square_list(lst):
    if not lst:
        return []
    
    return [num ** 2 for num in lst]
print(square_list([1, 2, 3, 4]))
```
