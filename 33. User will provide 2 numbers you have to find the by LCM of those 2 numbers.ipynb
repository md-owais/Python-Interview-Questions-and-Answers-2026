# 33.	User will provide 2 numbers you have to find the by LCM of those 2 numbers

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

while a % b != 0:
    rem = a % b
    a = b
    b = rem
hcf = b

lcm = (a*b)/hcf
print('LCM is:', lcm)
```

```python
# Best one for interview
# This is the clean and efficient version:
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

a, b = num1, num2

while b != 0:
    a, b = b, a % b

lcm = (num1 * num2) // a
print("LCM is:", lcm)
```

```python
# Using math module
import math

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

lcm = (num1 * num2) // math.gcd(num1, num2)
print("LCM is:", lcm)
```

```python
# Using Loop
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

max_num = max(num1, num2)

while True:
    if max_num % num1 == 0 and max_num % num2 == 0:
        print("LCM is:", max_num)
        break
    max_num += 1
```
