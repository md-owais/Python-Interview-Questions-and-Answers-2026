# 42.	Write a program to print the following pattern

```python
n = int(input("Enter peak number: "))

# Inreasing part
for i in range(1, n + 1): # 1 to peak number
    print("*" * i)  # print i stars

# Decreasing part
for i in range(n - 1, 0, -1): # peak number to 1
    print("*" * i) # print i stars
```

```python
# using nested loops
n = 3

# Increasing
for i in range(1, n + 1):
    for j in range(i):
        print("*", end="")
    print()

# Decreasing
for i in range(n - 1, 0, -1):
    for j in range(i):
        print("*", end="")
    print()
```
