# 52.	Extract username from a given email. 

Eg if the email is nitish24singh@gmail.com then the username should be nitish24singh


```python
email = input("Enter your email: ")
username = email.split('@')[0] # split at '@' and take first part
print(username)
```

```python
# Using loop (without split)
email = input("Enter email: ")

username = ""

for ch in email:
    if ch == "@": # stop when '@' is found
        break
    username += ch # build username
    
print("Username:", username)
```

```python
# Using index()
email = input("Enter email: ")

pos = email.index("@")   # find position of '@'

username = email[:pos]   # slice till that position

print("Username:", username)
```
