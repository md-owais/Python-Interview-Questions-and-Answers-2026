# 55. Count the number of vowels in a string provided by the user.

```python
a = input("Enter your string: ")
vowels = 'aeiou'
count = 0

for i in a:
     if i in vowels:
            count = count + 1
            
print(count)
```

```python
# Using loop (Best for interviews)
s = input("Enter a string: ")

count = 0
vowels = "aeiouAEIOU"   # includes both lowercase and uppercase

for ch in s:            # iterate through each character
    if ch in vowels:    # check if character is a vowel
        count += 1      # increase count

print("Number of vowels:", count)
```

```python
# Using lower() (clean approach)
s = input("Enter string: ").lower()

count = 0

for ch in s:
    if ch in "aeiou":
        count += 1

print("Vowels:", count)
```

```python
# Using sum() (Pythonic way)
s = input("Enter string: ")

count = sum(1 for ch in s if ch.lower() in "aeiou")

print("Vowels:", count)
```

```python
# Count each vowel separately
s = input("Enter string: ").lower()

vowel_count = {"a":0, "e":0, "i":0, "o":0, "u":0}

for ch in s:
    if ch in vowel_count:
        vowel_count[ch] += 1

print(vowel_count)
```
