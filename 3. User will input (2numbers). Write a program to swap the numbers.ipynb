3.	User will input (2numbers).Write a program to swap the numbers

```python
# Using Temporary Varialble
a = int(input("Enter value of a: "))
b = int(input("Enter value of b: "))
temp = a
a = b
b = temp
print("After swapping:")
print("Value of a:", a)
print("Value of b:", b)
```

```python
# Without Using Temporary Varialble
a = int(input("Enter value of a: "))
b = int(input("Enter value of b: "))

a, b = b, a

print("After swapping:")
print("Value of a:", a)
print("Value of b:", b)
```

```python
# using Addition and Subtraction
a = int(input("Enter value of a: "))
b = int(input("Enter value of b: "))

a = a + b
b = a - b
a = a - b

print("After swapping:")
print("Value of a:", a)
print("Value of b:", b)
```

```python
# Using Multiplication and Division - Works only if numbers are not zero
a = int(input("Enter value of a: "))
b = int(input("Enter value of b: "))

a = a * b
b = a / b
a = a / b

print("After swapping:")
print("Value of a:", a)
print("Value of b:", b)
```

```python
# Using XOR (Bitwise Operation)
a = int(input("Enter value of a: "))
b = int(input("Enter value of b: "))

a = a ^ b
b = a ^ b
a = a ^ b

print("After swapping:")
print("Value of a:", a)
print("Value of b:", b)
```
