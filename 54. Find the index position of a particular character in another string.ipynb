# 54.	Find the index position of a particular character in another string. 

```python
a = input("Enter String: ")
b = input("Enter Character: ")
print(a.index(b))
```

```python
s = "hello how are you"
char = 'h'

position = s.index(char)
print("Index:", position)
```

```python
# Using find() method (Safer)
s = "hello how are you"
char = 'h'

position = s.find(char)
print("Index:",position)
```

```python
# Find all index positions (Important Interview Variation)
s = "hello how are you"
char = 'h'

positions = []

for i in range(len(s)):
    if s[i] == char:
        positions.append(i)
print("All positions:", positions)
```

```python
# Using list comprehension (Pythonic way)
s = "hello how are you"
char = 'h'

positions = [i for i, c in enumerate(s) if c == char]
print("All positions:", positions)
```
