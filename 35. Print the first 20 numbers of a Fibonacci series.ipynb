# 35.	Print the first 20 numbers of a Fibonacci series

```python
a, b = 0, 1

for i in range(20):
    print(a)
    a, b = b, a+b
```

```python
n = 20

a, b = 0, 1

for i in range(n):
    print(a)
    a, b = b, a + b
```

```python
# One-line style (same logic)
a, b = 0, 1
for _ in range(20):
    print(a, end=" ")
    a, b = b, a + b
```
