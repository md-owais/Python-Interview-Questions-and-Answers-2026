```python
# Using nested loops (Best for logic)
n = 5
for i in range(1, n + 1):
    # Increasing part (1 -> i)
    for j in range(1, i + 1):
        print(j, end= ' ')
        
    # Decreasing part (i-1 -> 1)
    for j in range(i - 1, 0, -1):
        print(j, end= " ")
    print()
```

# Method 2: Cleaner version

```python
n = 5

for i in range(1, n + 1):
    inc = list(range(1, i + 1))
    dec = list(range(i - 1, 0, -1))
    
    row = inc + dec
    print(*row)
```
