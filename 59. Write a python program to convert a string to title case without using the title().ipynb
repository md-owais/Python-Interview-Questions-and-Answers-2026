# 59.	Write a python program to convert a string to title case without using the title()

```python
a = 'hEllo oWais'
y = a.split()
r = ''

for i in y:
    r = r + i[0].upper() + i[1:].lower() + ' '
print(r)  
```

```python
# 1. Using built-in title()
def to_title_case(s):
    return s.title()

# Example
text = "hello world from python"
print(to_title_case(text))
```

```python
# 2. WITHOUT using title() (Manual Logic – Interview Important ⭐)
def to_title_case(s):
    words = s.split()          # Split sentence into words
    result = []
    
    for word in words:
        # Capitalize first letter + keep rest as lowercase
        result.append(word[0].upper() + word[1:].lower())
    
    return " ".join(result)

# Example
text = "hello world from python"
print(to_title_case(text))
```

```python
# Edge Case Handling (Better Version 💡)
def to_title_case(s):
    words = s.split()
    result = []
    
    for word in words:
        if word:  # avoid empty strings
            result.append(word[0].upper() + word[1:].lower())
    
    return " ".join(result)

text = "hello world from python"
print(to_title_case(text))
```
