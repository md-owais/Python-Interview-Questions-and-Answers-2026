# 67.	Create 2 lists from a given list where 1st list will contain all the odd numbers from the original list and the 2nd one will contain all the even numbers 

```python
L = [1,2,3,4,5,6,7,8,9]
L_even = []
L_odd = []

for i in L:
    if i % 2 == 0:
        L_even.append(i)
    else:
        L_odd.append(i)
print('List of even values:', L_even)
print('List of odd values:', L_odd)
```

```python
# 1. Using Loop (Interview Standard ⭐)
def split_odd_even(lst):
    odd = []
    even = []
    
    for num in lst:
        if num % 2 == 0:
            even.append(num)
        else:
            odd.append(num)
    
    return odd, even

# Example
data = [1, 2, 3, 4, 5, 6]
odd_list, even_list = split_odd_even(data)

print("Odd:", odd_list)
print("Even:", even_list)
```

```python
# 2. Using List Comprehension (Pythonic ⭐)
def split_odd_even(lst):
    odd = [num for num in lst if num % 2 != 0]
    even = [num for num in lst if num % 2 == 0]
    return odd, even

data = [1, 2, 3, 4, 5, 6]
odd, even = split_odd_even(data)

print("Odd:", odd)
print("Even:", even)
```
