```python
n = 5
for i in range(1, n + 1):
    for j in range(n - i): # print spaces for alignment
        print(" ", end = " ")
        
    # print stars (2*i - 1 stars)
    for k in range(2 * i - 1):
        print("*", end = " ")
    print() # move to next line
```

```python
# Method 2: Cleaner Python way
n = 5

for i in range(1, n + 1):
    print(" " * (n - i) + "*" * (2 * i - 1))
```
