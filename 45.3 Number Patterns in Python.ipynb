```python
for i in range(1, 5):
    for j in range(1, i+1):
        print(j, end="")
    print()
```

```python
for i in range(4, 0, -1):
    for j in range(1, i+1):
        print(j, end="")
    print()
```

```python
for i in range(1, 5):
    
    for j in range(1, i+1):
        print(j, end="")
    for j in range(i-1, 0, -1):
        print(j, end="")
    print()  
```

```python
num = 1

for i in range(1, 5):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```
