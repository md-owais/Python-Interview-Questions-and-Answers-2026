# 40. Find the reverse of a number provided by the user(any number of digit) 

```python
num = input("Enter a number: ")
result = num[::-1]
print("The reversed number is:", result)
```

```python
print(int(str(input("Enter number: "))[::-1]))
# This won’t handle negative numbers properly
```

```python
# Using loop (Best for logic)
num = int(input("Enter a number: "))

rev = 0
temp = abs(num)

while temp > 0:
    digit = temp % 10
    rev = rev * 10 + digit
    temp //= 10

# Handle negative numbers
if num < 0:
    rev = -rev

print("Reversed number:", rev)
```

```python
# Using string (Easy way)
num = input("Enter a number: ")

if num[0] == '-':
    rev = '-' + num[:0:-1]
else:
    rev = num[::-1]

print("Reversed number:", rev)
```
