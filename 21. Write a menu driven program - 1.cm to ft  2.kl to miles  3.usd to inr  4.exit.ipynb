# 21.	Write a menu driven program - 1.cm to ft  2.kl to miles  3.usd to inr  4.exit

```python
user_input = input('''
Hi! Welcome to my page
What would you like to do?

1. Convert cm to inches
2. Convert km to miles
3. Convert usd to inr
4. Exit
''')

if user_input == '1':
    cm = float(input("Enter value in cm : "))
    inch = cm * 30.48
    print('Your value in inch is-', inch)
elif user_input == '2':
    km = float(input('Enter value in km: '))
    mile = km * 0.621
    print('Your value in miles is-', mile)
elif user_input == '3':
    usd = float(input("Enter value in usd: "))
    inr = usd * 83
    print('Your value in inr is-', inr)
else:
    print('Exit')
```

```python
while True:
    print("\n----- MENU -----")
    print("1. Convert cm to feet")
    print("2. Convert km to miles")
    print("3. Convert USD to INR")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        cm = float(input("Enter value in cm: "))
        feet = cm / 30.48
        print("In feet:", feet)

    elif choice == 2:
        km = float(input("Enter value in km: "))
        miles = km * 0.621371
        print("In miles:", miles)

    elif choice == 3:
        usd = float(input("Enter amount in USD: "))
        inr = usd * 83   # Approx conversion rate
        print("In INR:", inr)

    elif choice == 4:
        print("Exiting program...")
        break

    else:
        print("Invalid choice! Please try again.")
```
