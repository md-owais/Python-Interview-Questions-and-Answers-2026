# 22.	Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

```python
heads = int(input("Enter total heads: "))
legs = int(input("Enter total legs: "))

# Solve using formula
dogs = (legs - 2 * heads) // 2
chickens = heads - dogs

# Validation
if dogs < 0 or chickens < 0 or (4*dogs + 2*chickens != legs):
    print("Invalid input!")
else:
    print("Dogs:", dogs)
    print("Chickens:", chickens)
```

```python
heads = int(input())
legs = int(input())

for d in range(heads + 1):
    c = heads - d
    if 4*d + 2*c == legs:
        print("Dogs:", d, "chickens:", c)
        break
else:
    print("Invalid Input")
```
