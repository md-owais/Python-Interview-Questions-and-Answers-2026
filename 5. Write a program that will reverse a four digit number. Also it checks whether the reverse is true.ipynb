5.	Write a program that will reverse a four digit number. Also it checks whether the reverse is true.

```python
# User Input
user_input = int(input("Enter your four digit number: "))
num = user_input
a = num % 10
num = num // 10

b = num % 10
num = num // 10

c = num % 10
d = num // 10
# formula
rev = 1000*a + 100*b + 10*c +d
print('Original num:', user_input)
print('reverse num:', rev)
# check
if user_input == rev:
    print('True')
else:
    print('False')
```

```python
num = int(input("Enter a four digit number: "))

original = num

a = num % 10
num = num // 10

b = num % 10
num = num // 10

c = num % 10
d = num // 10

reverse = a*1000 + b*100 + c*10 + d

print("Reversed number:", reverse)

if original == reverse:
    print("The number is same after reverse (Palindrome)")
else:
    print("The number is not same after reverse")
```

# A palindrome is a value that reads the same forward and backward.
Examples: 121, madam, level

```python
# Palindrome Number (Using Reverse Logic)
num = int(input('Enter number: '))
original = num
reverse = 0
while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10
if original == reverse:
    print("Palindrome number")
else:
    print("Not a Palindrome")
```

```python
# Palindrome Number (Using String)
num = input("Enter Number: ")
if num == num[::-1]:
    print("Palindrome number")
else:
    print("Not a Palindrome")
```

```python
# Palindrome String
text = input("Enter a string: ")

if text == text[::-1]:
    print("Palindrome string")
else:
    print("Not palindrome")
```

```python
# Palindrome Using For Loop
text = input("Enter a string: ")
reverse = ""
for i in text:
    reverse = i + reverse
if text == reverse:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# Palindrome Using While Loop
text = input("Enter string: ")

i = 0
j = len(text) - 1
flag = True

while i < j:
    if text[i] != text[j]:
        flag = False
        break
    i += 1
    j -= 1

if flag:
    print("Palindrome")
else:
    print("Not palindrome")
```

```python
# Palindrome Function
def palindrome(s):
    return s == s[::-1]
value = input("Enter value: ")
if palindrome(value):
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# Palindrome for List
lst = [1,2,3,2,1]
if lst == lst[::-1]:
    print("Palindrome list")
else:
    print("Not Palindrome")
```
