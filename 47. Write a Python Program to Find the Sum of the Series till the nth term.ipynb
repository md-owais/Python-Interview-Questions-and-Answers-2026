# 47. Write a Python Program to Find the Sum of the Series till the nth term: 
1 + x^2/2 + x^3/3 + … x^n/n

n will be provided by the user


```python
x = int(input("Enter value of x: "))
n = int(input("Enter value of n: "))

sum_series = 1   # first term is 1

for i in range(2, n + 1):
    term = (x ** i) / i     # calculate x^i / i
    sum_series += term      # add to total

print("Sum of series:", sum_series)
```

```python
#Method 2: Using while loop
x = int(input("Enter x: "))
n = int(input("Enter n: "))

sum_series = 1
i = 2

while i <= n:
    sum_series += (x ** i) / i
    i += 1

print("Sum:", sum_series)
```

```python
# Optimized Idea (Advanced)
#Instead of calculating x^i every time:

x = int(input("Enter x: "))
n = int(input("Enter n: "))

sum_series = 1
power = x   # stores x^i

for i in range(2, n + 1):
    power *= x               # builds x^i step by step
    sum_series += power / i

print("Sum:", sum_series)
```
