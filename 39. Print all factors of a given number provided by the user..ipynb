# 39.	Print all factors of a given number provided by the user.

```python
num = int(input("Enter a number: "))
for i in range(1, num + 1):
    if num % i == 0:
        print("The factors is:", i)
```

```python
# Optimized Method (Better for interviews)
num = int(input("Enter a number: "))

print("Factors are:")

for i in range(1, int(num**0.5) + 1):
    if num % i == 0:
        print(i)
        if i != num // i:
            print(num // i)
```
