# 65. Write a program that can count the number of words in a given string

```python
a = input('Enter your string: ')
count = 0

for i in a:
    if i == ' ':
        count = count + 1
words = count + 1
print('Number of words in the string:', words)
```

# 1. Using split() (Most Common)

```python
def count_words(s):
    return len(s.split())
text = "Hello world from Python"
print(count_words(text))
```

# 2. Using re (Handles Multiple Spaces & Symbols)

```python
import re

def count_words(s):
    words = re.findall(r'\b\w+\b', s)
    return len(words)

# Example
print(count_words("Hello, world! Python is great."))
```

# 4. Handle Edge Cases (Better Version)

```python
def count_words(s):
    if not s.strip():
        return 0
    return len(s.split())

print(count_words("Hello, world! Python is great."))
```
