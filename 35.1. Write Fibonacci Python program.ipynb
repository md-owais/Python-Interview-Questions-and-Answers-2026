```python
# Using Loop (Best for interviews)
n = int(input("Enter numer of terms: "))

a, b = 0, 1

for i in range(n):
    print(a)
    a, b = b, a+b
```

```python
# Using while loop
n = int(input("Enter number of terms: "))

a, b = 0, 1
count = 0

while count < n:
    print(a)
    a, b = b, a + b
    count += 1
```

```python
# Using recursion (Important concept)
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

n = int(input("Enter number of terms: "))

for i in range(n):
    print(fib(i))
```

```python
# Store in List
n = int(input())

fib = [0, 1]

for i in range(2, n):
    fib.append(fib[i-1] + fib[i-2])

print(fib[:n])
```
