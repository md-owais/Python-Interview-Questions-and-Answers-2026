# 29.	Print all the armstrong numbers in the range of 100 to 1000

```python
for num in range(100, 1000):
    i = num
    a = num % 10
    num = num // 10
    b = num % 10
    c = num // 10
    
    if (a**3) + (b**3) + (c**3) == i:
        print(i)
    i += 1
```

```python
for num in range(100, 1000):
    temp = num
    total = 0

    while temp > 0:
        digit = temp % 10
        total += digit ** 3
        temp = temp // 10

    if total == num:
        print(num)
```

```python
for num in range(100, 1000):
    total = sum(int(digit) ** 3 for digit in str(num))
    if total == num:
        print(num)
```
