# 58. Write a python program to remove all the duplicates from a list

```python
L1 = [1,2,2,3,4,4,5,6,7,7]
L2 = []

for i in L1:
    if i not in L2:
        L2.append(i)
print(L2)
```

```python
# Using set() (Fastest & Simple)
def remove_duplicates(lst):
    return list(set(lst))

# Example
data = [1, 2, 2, 3, 4, 4, 5]
print(remove_duplicates(data))
```

```python
# Preserve Order (Best for Interviews ⭐)
def remove_duplicates(lst):
    unique = []
    for item in lst:
        if item not in unique:
            unique.append(item)
    return unique

# Example
data = [1, 2, 2, 3, 4, 4, 5]
print(remove_duplicates(data))
```

```python
# Using dict.fromkeys() (Best Pythonic Way)
def remove_duplicates(lst):
    return list(dict.fromkeys(lst))

# Example
data = [1, 2, 2, 3, 4, 4, 5]
print(remove_duplicates(data))
```

```python
# Using Set + Loop (Optimized Order + Performance)
def remove_duplicates(lst):
    seen = set()
    result = []
    
    for item in lst:
        if item not in seen:
            seen.add(item)
            result.append(item)
    
    return result
data = [1, 2, 2, 3, 4, 4, 5]
print(remove_duplicates(data))
```
