# 53. Count the frequency of a particular character in a provided string. Eg 'hello how are you' is the string, the frequency of h in this string is 2.

```python
a = input("Enter String: ")
b = input("Enter the character: ")
count = 0

for i in a:
    if i in b:
        count = count + 1
print("Character occured:",count, 'times')
```

```python
# Using built-in count() method (Best & simplest)
s = 'hello how are you'
char = 'h'

frequency = s.count(char)
print("Frequency:", frequency)
```

```python
# Using loop (without built-in functions)
s = "hello how are you"
char = 'h'

count = 0
for c in s:
    if c == char:
        count += 1
print("Frequency:", count)

# Important when interviewer says “don’t use built-in functions”
```

```python
# Using dictionary (for all character frequencies)
s = "hello how are you"

freq = {}

for c in s:
    if c in freq:
        freq[c] += 1
    else:
        freq[c] = 1
        
print(freq)
print("Frequency of h:", freq.get('h', 0))

#Useful when you need frequency of all characters
```

```python
# Using collections.Counter (Advanced & Pythonic)
from collections import Counter

s = "hello how are you"

freq = Counter(s)
print(freq)
print("Frequency of h:", freq['h'])
```
