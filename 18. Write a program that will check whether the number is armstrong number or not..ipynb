# 18.	Write a program that will check whether the number is armstrong number or not.

An Armstrong number (also called a narcissistic number) is a number that is equal to the sum of its digits each raised to the power of the number of digits.

Example:

153

1^3 + 5^3 + 3^3 = 1 + 25 + 27 = 153

```python
user_input = int(input("Enter your number: "))
num = user_input

a = num % 10
num = num // 10

b = num % 10
c = num // 10

if (a**3) + (b**3) + (c**3) == user_input:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

```python
num = int(input("Enter a number: "))

temp = num
n = len(str(num))   # Count number of digits
result = 0

while temp > 0:
    digit = temp % 10
    result += digit ** n
    temp = temp // 10

if result == num:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

```python
num = input("Enter a number: ")

n = len(num)
result = sum(int(digit) ** n for digit in num)

if result == int(num):
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")  
```
