# 1. Reverse Number (Loop – Best for interviews)

```python
num = int(input("Enter number: "))

rev = 0  # This will store the reversed number
temp = abs(num) # abs() ensures negative numbers become positive for process

while temp > 0: # Loop runs untill all digits are processed
    digit = temp % 10 # Extract last digit (ex - 123 ->3)
    rev = rev * 10 + digit # shifts left and add digit (ex- 12 -> 123)
    temp //= 10 # Remove last digit (123 -> 12)

# If original number was negative, restore the sign
if num < 0:
    rev = -rev
print("Reversed:", rev)
```

# 2. Reverse Number (String Method)

```python
num = input("Enter number: ")

# check if number is negative
if num[0] == '-':
    rev = '-' + num[:0:-1] # skips '-' and reverse the string
else:
    rev = num[::-1] # [::-1] reverses the strings
    
print("Reversed:", rev)
```

# 3. One-liner Reverse

```python
# input -> convert to string -> reverse -> convert back to int
print(int(str(input("Enter number: "))[::-1])) # This dosen't handle negative numbers properly
```

```python
# Reverse Using Recursion
def reverse(num, rev =0):
    if num == 0:  # Base case: when number becomes 0
        return rev
    return reverse(num//10, rev*10 + num%10)
    # num//10 removes last digit
    # rev*10 + num%10 builds reversed number
num = int(input("Enter number: "))
print("Reversed:", reverse(abs(num)))
```

# 6. Reversed Words in Sentence

```python
s = input("Enter sentence: ")
rev = s[::-1]
print("Reversed:", rev)
```

```python
s = input("Enter sentence: ")

words = s.split()        # split sentence into words list
rev = " ".join(words[::-1])   # reverse list and join with space

print("Reversed sentence:", rev)
```

# 7. Reverse Each Word (not sentence order)

```python
s = input("Enter sentence: ")
# Reverse each word individually
rev = " ".join(word[::-1] for word in s.split())
print("Reversed:", rev)
```

# 9. Reverse List (Manual Loop)

```python
lst = [1, 2, 3, 4, 5]

rev = []

# Loop from last index to first
for i in range(len(lst)-1, -1, -1):
    rev.append(lst[i]) # add elements in reverse order
print(rev)
```

# 12. Reverse Number (Simplified Loop)

```python
num = int(input("Enter number: "))
rev = 0

while num:              # runs until num becomes 0
    rev = rev * 10 + num % 10
    num //= 10

print(rev)
```

# 13. Store Reversed Digits in List

```python
num = int(input())

digits = []

while num:
    digits.append(num % 10)  # add last digit to list
    num //= 10

print(digits)
```

# 14. Reverse + Palindrome Check

```python
num = int(input("Enter number: "))
temp = num     # store original number
rev = 0

while num > 0:
    rev = rev * 10 + num % 10
    num //= 10

print("Reversed:", rev)

# Compare original and reversed
if temp == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

# 15. Reverse Float Number

```python
num = input("Enter float: ")

# Check if decimal exists
if '.' in num:
    int_part, dec_part = num.split('.')   # split integer and decimal
    rev = int_part[::-1] + '.' + dec_part[::-1]
else:
    rev = num[::-1]

print("Reversed:", rev)
```
