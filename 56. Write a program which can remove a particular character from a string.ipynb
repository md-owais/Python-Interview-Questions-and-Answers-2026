# 56. Write a program which can remove a particular character from a string. 

```python
a = input("Enter String: ")
b = int(input("Enter the value of the character: "))
c = a[0:b-1]
d = a[b:]
print(c+d)
```

```python
# Using loop (Best for interviews)
s = input("Enter string: ")
ch = input("Enter character to remove: ")

result = ""

for c in s:                  # iterate each character
    if c != ch:              # keep only if not equal
        result += c          # build new string

print("Result:", result)
```

```python
# Using replace() (Shortcut)
s = input("Enter string: ")
ch = input("Enter character to remove: ")

result = s.replace(ch, "")   # replace with empty string

print("Result:", result)
```

```python
# Remove only first occurrence
s = input("Enter string: ")
ch = input("Enter character: ")

pos = s.find(ch)    # find first occurrence

if pos != -1:
    result = s[:pos] + s[pos+1:]
else:
    result = s

print("Result:", result)
```
